```python
cols = []
for i in range(len(file1.columns)):
    cols.append(file1.columns[i].lower())
file1.columns = cols

cols1 = []
for i in range(len(file2.columns)):
    cols1.append(file2.columns[i].lower())
file2.columns = cols1

cols2 = []
for i in range(len(file3.columns)):
    cols2.append(file3.columns[i].lower())
file3.columns = cols2

display(file1.head()) 
display(file2.head()) 
display(file3.head()) 
```

```python
column_names = file2.columns
column_names
file2 = file2[['customer', 'st', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
file2.head

#column_names = file3.columns
#column_names
#file3 = file3[['customer', 'st', 'gender', 'education', 'customer lifetime value',
#       'income', 'monthly premium auto', 'number of open complaints',
#       'policy type', 'vehicle class', 'total claim amount']]
#display(file3.head()) 
```

```python
column_names = file3.columns
column_names
file3 = file3[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
file3.head
```

```python
data = pd.concat([file1, file2, file3], axis=0) #concatination
data.shape
#data.info()
print(data._get_numeric_data()) #numerical
data.select_dtypes(include=['object']).columns #whih columns are categorical

```

```python
#Delete the column education and the number of open complaints from the dataframe.

data1=data.drop(['education', 'number of open complaints'],axis=1)
display(data1.head())
```

```python
#Correct the values in the column customer lifetime value. 
#They are given as a percent, so multiply them by 100 and change dtype to numerical type.

data['customer lifetime value'] = pd.to_numeric(data['customer lifetime value'], errors='coerce')#convert to numeric
                                                
data['customer lifetime value']=data['customer lifetime value']*100 #multip 100

display(data.tail())
```

```python
#Check for duplicate rows in the data and remove if any.
data = data.drop_duplicates()
data.shape # it's now 3 lines less
```

```python
#Filter out the data for customers who have an income of 0 or less.

data1= data[data['income']>0]
display(data1.head())
```
